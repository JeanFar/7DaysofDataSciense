```python
# Criando um novo data frame que indique a quantidade de emprestimos por data e alterando o tipo de data para datetime
emprestimos_data= pd.DataFrame(planilha_completa['data_emprestimo'].value_counts()).reset_index()
emprestimos_data.columns= ['data', 'quantidade']
emprestimos_data['data'] = pd.to_datetime(emprestimos_data['data'])
emprestimos_data
```

```python
emprestimos_p_ano= emprestimos_data.groupby(by=emprestimos_data.data.dt.year).sum(numeric_only=True)
emprestimos_p_ano.index.name= 'ano'
emprestimos_p_ano
```

```python
# Repare que só temos os dados do primeiro trimestre de 2020
emprestimos_por_trimestre= emprestimos_data['data'].dt.to_period('Q').astype(str)
emprestimos_por_trimestre = emprestimos_por_trimestre.str[:4] + '-' + emprestimos_por_trimestre.str[-1]
emprestimos_por_trimestre = emprestimos_por_trimestre.value_counts().sort_index()
emprestimos_por_trimestre = emprestimos_por_trimestre.reset_index()
emprestimos_por_trimestre.columns = ['trimestre', 'quantidade']
```

```python
print(emprestimos_por_trimestre)
```

```python
emprestimos_por_semestre= {
    'semestre':['2010.1', '2010.2', '2011.1', '2011.2', '2012.1', '2012.2', '2013.1', '2013.2', '2014.1', '2014.2', '2015.1', '2015.2', '2016.1', '2016.2', '2017.1', '2017.2', '2018.1', '2018.2', '2019.1', '2019.2', '2020.1'],
    'quantidade': [79350, 81927, 89962, 106237, 116601, 104540, 121291, 113140, 105171, 113594, 119755, 93362, 112778, 102633, 112980, 104994, 102523, 91024, 91289, 80007, 25533]
}
emprestimos_por_semestre= pd.DataFrame(emprestimos_por_semestre)

emprestimos_por_semestre

```

```python
plt.figure(figsize=(12, 6))
plot_linha_data= sns.lineplot(data=emprestimos_por_semestre, x='semestre', y='quantidade')
plot_linha_data.set(xlabel=None, ylabel=None)
plot_linha_data.tick_params(axis='x', rotation=30)
plot_linha_data.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
plot_linha_data.set_title('Quantidade de empréstimos feitos por Trimestre'+'\n',size=20,loc='left',weight='bold')
sns.barplot(data=emprestimos_por_semestre, x='semestre', y='quantidade', alpha=0.2)
plt.show()
```

```python
# Vemos aqui que tivemos um bom crescimento no número de empréstimos ao longo dos anos, com eles sempre se mantendo altos, acima de 90000 ao longo de 2011.2 até 2019.1;
# A partir de 2019.2 tivemos uma queda expressiva, abaixo inclusive dos primeiros dados que tivemos em 2010.1;
# Se cruzarmos com nosso dataframe de trimestres , em todos os primeiros trimestres dos anos, alcançava-se mais de 90 mil emprestimos, enquanto que no primeiro trimestre de 2020.1 tivemos 25533 emprestimos.
# Podemos relacionar esses números com a pandemia do Covid-19 que tivemos no início de 2020, logo em meados de março, mas ainda com alguma dúvida, pois a maioria das nossas cidades só decretou lockdown em março.

# Temos picos em 2013.1 e 2015.1, e isso nos indica benchmarks, que quando cruzados com as estratégias utilizadas da conjuntura desses anos, podemos mapear possíveis ações para alcançar padrões semelhantes no contexto atual

```

```python
# A diretoria também gostaria de gerenciar melhor os recursos humanos da biblioteca de acordo com a demanda de trabalho existente. Por exemplo:


# gerenciar a programação de férias dos colaboradores de acordo com os meses de menor demanda;
# programar atividades que não sejam de atendimento ao usuário para períodos específicos de menor demanda.

# Há uma suspeita interna de que os meses com maior número de exemplares emprestados sejam março e setembro, mas não foi realizada uma análise real sobre isso.

# Portanto, gere uma tabela com a quantidade total de exemplares emprestados por mês
# e descubra quais meses são os que possuem a maior quantidade de empréstimos realizados. Plote um gráfico de linhas.
```

```python
# id_exemplar
emprestimos_por_mes= emprestimos_data.groupby(by= emprestimos_data.data.dt.month).sum(True)
emprestimos_por_mes.index.name='mes'
emprestimos_por_mes
dicionario_mes= {1:'Jan', 2:'Fev', 3:'Mar', 4: 'Abr', 5: 'Mai', 6: 'Jun', 7:'Jul', 8: 'Ago', 9: 'Set', 10:'Out', 11:'Nov', 12:'Dez'}
emprestimos_por_mes.index= emprestimos_por_mes.index.map(dicionario_mes)
emprestimos_por_mes
```

```python
plt.figure(figsize=(12, 6))
plot_emprestimos_mes = sns.lineplot(data=emprestimos_por_mes, x='mes', y='quantidade')
plot_emprestimos_mes.set(xlabel=None, ylabel=None)
plot_emprestimos_mes.tick_params(axis='x', rotation=30)
plot_emprestimos_mes.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',', '.')))
plot_emprestimos_mes.set_title('Quantidade de empréstimos feitos por Mês' + '\n', size=20, loc='left', weight='bold')
```

```python
# Vemos que temos pontos de pico na quantidade de empréstimos de Fevereiro a Maio e de Agosto a outubro. Seguindo as suspeitas internas.
# Podemos ver espaço para que os funcionários tirem férias em Janeiro, Junho, Julho, Novembro e Dezembro.
# Sendo que os meses com a menor demanda são Janeiro e dezembro, o que nos indica possibilidades de que mais funcionários tirem férias nesse período, e inclusive minimizando custos para a biblioteca.
```

```python
# Quero excluir os outliers
emprestimos_por_hora= emprestimos_data.groupby(by=emprestimos_data.data.dt.hour).sum(True)
emprestimos_por_hora.index.name='hora'
emprestimos_por_hora= emprestimos_por_hora.reset_index()
emprestimos_por_hora
```

```python
emprestimos_por_hora_limpo= emprestimos_por_hora.drop([0, 1, 17, 18]).reset_index(drop=True)

emprestimos_por_hora_limpo
```

```python
plt.figure(figsize=(12, 6))
plot_emprestimos_por_hora = sns.lineplot(data=emprestimos_por_hora_limpo, x='hora', y='quantidade')
plot_emprestimos_por_hora.set(xlabel=None, ylabel=None)
plot_emprestimos_por_hora.tick_params(axis='x', rotation=30)
plot_emprestimos_por_hora.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',', '.')))
plot_emprestimos_por_hora.set_title('Quantidade de empréstimos feitos por hora' + '\n', size=20, loc='left', weight='bold')
```
